# Explore Shape

```python
import h5py
from pathlib import Path
import numpy as np

base_dir = Path("./data/ToM_Data")
subject_dir = base_dir / "Human_fMRI_Data" / "Belief_Period_H5" / "TOM001"
file_names = ["Left_PFC.h5", "Left_TPJ.h5", "Right_PFC.h5", "Right_TPJ.h5"]

# 最终数据结构: data[region][condition][trial][dataset_name]
data = {}

for fname in file_names:
    file_path = subject_dir / fname
    region_name = fname.replace(".h5", "")
    data[region_name] = {}

    with h5py.File(file_path, 'r') as f:
        # L1: 遍历 Conditions (e.g., 'False_Belief')
        for condition_key, condition_group in f.items():
            data[region_name][condition_key] = {}

            # L2: 遍历 Trials (e.g., 'trial_00')
            for trial_key, trial_group in condition_group.items():
                data[region_name][condition_key][trial_key] = {}

                # L3: 遍历 Datasets (e.g., 'Signal', 'Story')
                for dataset_key, dataset in trial_group.items():
                    data[region_name][condition_key][trial_key][dataset_key] = dataset[()]

# 访问示例: signal_data = data['Left_PFC']['False_Belief']['trial_00']['Signal']
```

```python
# 定义要检查的样本路径
region_to_check = 'Left_PFC'
condition_to_check = 'False_Belief'
trial_to_check = 'trial_00'

print(f"--- 检查路径: [{region_to_check}][{condition_to_check}][{trial_to_check}] ---")

# --- Signal ---
sig = data[region_to_check][condition_to_check][trial_to_check]['Signal']
print(f"\n[Signal]")
print(f"  Shape: {sig.shape}")
print(f"  Type:  {sig.dtype}")
print(f"  Content (Head): \n{sig[:2, :5]}")  # 仅打印 2x5 子矩阵

# --- Condition_Name ---
name = data[region_to_check][condition_to_check][trial_to_check]['Condition_Name']
print(f"\n[Condition_Name]")
print(f"  Shape: (Scalar)")  # 它是 bytes, 没有 .shape
print(f"  Type:  {type(name)}")
print(f"  Content: {name.decode('utf-8')}")  # 解码 bytes 为 string

# --- Condition_Number ---
num = data[region_to_check][condition_to_check][trial_to_check]['Condition_Number']
print(f"\n[Condition_Number]")
print(f"  Shape: {num.shape}")  # Numpy 标量的 shape 是 ()
print(f"  Type:  {num.dtype}")
print(f"  Content: {num}")

# --- Story ---
story = data[region_to_check][condition_to_check][trial_to_check]['Story']
print(f"\n[Story]")
print(f"  Shape: {story.shape}")  # Numpy 标量的 shape 是 ()
print(f"  Type:  {story.dtype}")
print(f"  Content: {story}")
```
