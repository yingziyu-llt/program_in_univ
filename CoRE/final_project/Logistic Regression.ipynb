```python
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np
from pathlib import Path
import pickle

DATA_DIR = Path("data/ToM_Data/Human_fMRI_Data")
OUTPUT_DIR = Path("data/results_supcon_fixed")
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)


def calculate_decoding_accuracy(embeddings_dict):
    print(f"\n{'='*20} 评估：解码准确率 {'='*20}")

    results = {}

    for task_id, content in embeddings_dict.items():
        X = content['embeddings']  # (n_samples, hidden_dim)
        y = content['labels']  # (n_samples, )

        # 使用线性分类器 (Linear SVM 或 Logistic Regression)
        # 这里的 C=1.0 是正则化参数
        clf = LogisticRegression(max_iter=2000, C=1.0, solver='lbfgs', multi_class='auto')

        # 5折交叉验证 (5-Fold Cross Validation)
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')

        avg_acc = np.mean(scores)
        results[task_id] = avg_acc

        print(f"Task: {task_id:35s} | Accuracy: {avg_acc:.2%} (+/- {np.std(scores):.2%})")

    return results


with open(OUTPUT_DIR / "all_supcon_embeddings_fixed.pkl", 'rb') as f:
    final_embeddings = pickle.load(f)

acc_results = calculate_decoding_accuracy(final_embeddings)
```
