```python
import pyecharts.options as opts
from pyecharts.charts import Map, Pie, Grid,Timeline
from pyecharts import options as opts
import pandas as pd
```

```python
# 读取CSV并转换数据结构（参考网页3数据构造逻辑）
df = pd.read_csv('data.csv')
years = [col for col in df.columns if '年' in col]
data_dict = {
    year: [[prov, df.loc[i, year]] for i, prov in enumerate(df['地区'])]
    for year in years
}
```

```python
from pyecharts.charts import Map, Pie, Grid, Timeline
from pyecharts import options as opts

# 初始化时间轴（左右布局）
timeline = Timeline(
    init_opts=opts.InitOpts(
        width="1400px",  # 增加总宽度
        height="800px",
        theme="light"
    )
)
timeline.add_schema(
    orient="horizontal",
    play_interval=2000,
    is_auto_play=True,
    pos_bottom="5%",
    pos_left="20%",
    pos_right="20%"
)

for year in sorted(years):
    # 地图配置（左侧55%宽度）
    map_chart = (
        Map()
        .add("GDP（亿元）", data_dict[year], "china")
        .set_global_opts(
            title_opts=opts.TitleOpts(title=f"{year}GDP分布", pos_left="10%"),
            visualmap_opts=opts.VisualMapOpts(
    pos_left="8%",
    pos_right="50%",
    pos_top="50%",   # 原30%改为50%下移
    pos_bottom="10%", 
    max_=140000
)
        )
    )
    
    # 饼图配置（右侧35%宽度）
    # 原代码片段
    ratio_data_sorted = sorted(data_dict[year], key=lambda x: x[1], reverse=True)

    # 修改后代码（添加其他类）
    total_gdp = sum(v for _, v in data_dict[year])  # 计算全国GDP总和
    main_data = ratio_data_sorted[:10]  # 取前10大省份
    other_sum = total_gdp - sum(v for _, v in main_data)  # 计算其他省份总和
    final_data = main_data + [("其他", other_sum)]  # 组合新数据    
    pie_chart = (
        Pie()
        .add(
        series_name="占比",
        data_pair=final_data,  # 改用包含其他类的数据
        center=["85%", "50%"],
        radius=["10%", "20%"],
        color = [
    "#5470c6", "#91cc75", "#fac858", "#ee6666", 
    "#73c0de", "#9a60b4", "#3ba272", "#fc8452",
    "#ea7ccc", "#b6a8de", "#999999"  # 灰色用于「其他」类
]
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(
                formatter="{b}: {d}%",
                position="outside"
            )
        )
    )

    # 组合布局
    grid = Grid()
    grid.add(map_chart, grid_opts=opts.GridOpts(pos_left="5%", pos_right="30%"))
    grid.add(pie_chart, grid_opts=opts.GridOpts(pos_left="60%", pos_right="5%"))
    
    timeline.add(grid, year)
    
timeline.render("gdp_timeline.html")
```
